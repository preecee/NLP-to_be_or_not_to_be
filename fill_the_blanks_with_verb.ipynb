```python
#----FUNCTION DEFINITION-----------

#This function process the raw corpus and does following:
#--> conert to lowercase
#--> replace /n by  ' '
#--> replace multiple consecutive ' ' by single ' '
def process_corpus(corpus):
    x=corpus.casefold()
    txt=x.replace('\n', ' ')
    post_corpus=re.sub(r' {2,}', ' ', txt)
    return post_corpus
```

```python
corpus = open('corpus.txt', 'r', encoding='utf-8-sig').read()
post_corpus = process_corpus(corpus)
blob = TextBlob(post_corpus)
```

```python
look_for = ['am','are','were','was','is','been','being','be']
sen =[]
vrb =[]
for sentence in blob.sentences:
    sentence.string.replace
    found =0
    for v in look_for:
        if v in sentence:
            x=v
            y=sentence.string.replace(v,' ---- ',1)
            found =1
            break
    if found ==1:
        sen.append(y)
        vrb.append(x)
```

```python
train=pd.DataFrame({'Sentence':sen,'Verb':vrb})
train['Verb']=train['Verb'].astype("category")
train['Sentence']=train['Sentence'].astype("str")
```

```python
#create pipeline for vectorizing the text input and then creating model sequentially
x_train=train['Sentence']
y_train=train['Verb']

#pred_model=Pipeline([('vectorize',TfidfVectorizer(stop_words='english')),('predictive_model', MultinomialNB())])# this model did not have good performance
pred_model=Pipeline([('vectorize',TfidfVectorizer(ngram_range=(1, 2))),
                   ('predictive_model', GradientBoostingClassifier())])
pred_model.fit(x_train,y_train)
```
